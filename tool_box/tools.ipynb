# DataFrame to html

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.array([['御坂美琴', '超電磁砲', 'https://zh.wikipedia.org/wiki/%E5%BE%A1%E5%9D%82%E7%BE%8E%E7%90%B4'], \
                            ['上条當麻', '幻想殺手', 'https://zh.wikipedia.org/wiki/%E4%B8%8A%E6%9D%A1%E7%95%B6%E9%BA%BB'], \
                            ['茵蒂克絲', '禁書目錄', 'https://zh.wikipedia.org/wiki/%E8%8C%B5%E8%92%82%E5%85%8B%E7%B5%B2']]),
                   columns=['name', 'title', 'link'])
```

```python
def make_clickable(val):
    return '<a href="{}">{}</a>'.format(val,val)

def content(df):
    table = df[['name','title','link']].\
        rename(columns = {'name':'名稱', 'title':'稱號','link':'連結'}).\
        style.format({'連結':make_clickable}).\
        set_table_styles([{'selector': '.row_heading','props': [('display', 'none')]},{'selector': '.blank.level0','props': [('display', 'none')]}]).\
        set_properties(**{'background-color':'white','color':'black','border-color':'black','border-top':'solid','border-width':'1px'})\
        .render()
    html=f'''
    <html>
    <body>
    <h4>Dear 收件者 </h4> 
    <p>給收件者的話</p>
    <p></p>
    {table}
    <p></p>
    </body>
    </html> 
    '''
    return html
from IPython.core.display import display, HTML
display(HTML(content(df)))
```
